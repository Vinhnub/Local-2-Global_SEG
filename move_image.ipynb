```python
from pathlib import Path
import shutil

# Thư mục chứa ảnh (bao gồm các thư mục con)
src_dir = Path(r"E:\SEG_Dataset\paris_1")

# Thư mục đích
dst_dir = Path(r"E:\PythonFile\Project\Local-2-Global_SEG\data\datasets\rparis6k\jpg")
dst_dir.mkdir(parents=True, exist_ok=True)

# Các định dạng ảnh
image_exts = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}

count = 0

for file in src_dir.rglob("*"):
    if file.is_file() and file.suffix.lower() in image_exts:
        shutil.copy2(file, dst_dir / file.name)
        count += 1

print(f"Đã copy {count} ảnh.")
```
